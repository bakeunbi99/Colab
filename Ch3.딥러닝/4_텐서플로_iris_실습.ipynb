```python
"""
날짜 : 2021/10/26 16:42
내용 : 텐서플로 iris 딥러닝 실습하기
"""

import tensorflow as tf
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import mse
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
# 데이터 불러오기
df_iris = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/file/iris.csv')
df_iris
```

```python
# 학습데이터
iris_data = df_iris[['sepal.length', 'sepal.width', 'petal.length', 'petal.width']]
iris_label = df_iris['variety']
iris_label
```

```python
# 원-핫 인코딩(문자열로 된 라벨을 0, 1로 부호화)
encoder = LabelEncoder()
encoder.fit(iris_label)

transformed_label = encoder.transform(iris_label)
iris_label_cate = to_categorical(transformed_label)
# transformed_label
# iris_label_cate
```

```python
# 학습데이터, 학습라벨
train_data, test_data, train_label, test_label = train_test_split(iris_data, iris_label_cate)
train_data
```

```python
# 학습모델 설계
model = Sequential()
model.add(Dense(units=4, input_dim=4, activation='sigmoid')) # units는 개수 상관 없음. 성능만 좋으면 된다.
model.add(Dense(units=3, activation='sigmoid'))

model.summary()
```

```python
# 모델 설정
# loss='mean_squared_error' : 수치 예측
# loss='ctegorical_crossentropy' : 다중 분류
# loss='binary_crossentropy' : 이진



model.compile(loss='categorical_crossentropy', optimizer='Adam', metrics=['acc']) # 최적화 함수, loss 정의

# loss는 떨어지고 정확도(acc) 는 높아짐
```

```python
# 학습하기
model.fit(train_data, train_label, epochs=1000)
```

```python
# 평가하기
result = model.evaluate(test_data, test_label)
result
```
