```python
"""
날짜 : 2021/11/04 16:37
내용: 스마트카 모델 실습하기
"""

import tensorflow as tf
import pandas as pd
from keras.models import load_model
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
# 데이터 불러오기
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/file/CarDrivingIncidentInfo_Test.csv')
df
```

```python
# 학습 데이터, 라벨
df_train = df.iloc[:, :-1].values
df_label = df.iloc[:, -1].values # 뒤에서 부터 index

#X는 훈련데이터, Y는 라벨데이터
X_train, X_test, Y_train, Y_test = train_test_split(df_train, df_label, test_size=0.2, random_state=1)

Y_train = to_categorical(Y_train)
Y_test = to_categorical(Y_test)

sc = StandardScaler() # StandardScaler() 객체를 가지고 정규화 작업을 한다.
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
# 모델 불러오기
model = load_model('/content/drive/MyDrive/Colab Notebooks/model/smartcar_model.h5')
model
```

```python
# 모델 평가
result = model.evaluate(X_train, Y_train)
result
```

```python
# 테스트하기
pre = model.predict(X_test)
pre.round()
```
