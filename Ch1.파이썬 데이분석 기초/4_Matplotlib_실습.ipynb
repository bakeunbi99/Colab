```python
"""
날짜 : 2021/10/18 16:35
내용 : 파이썬 Matplotlib 실습하기

Matplotlib 라이브러리
  - 파이썬 데이터 시각화 패키지
  - Matplotlib를 기반으로 하는 seabon 시각화 라이브러리

가로 : 독립변수
세로 : 종속변수
"""

import matplotlib.pyplot as plt
import numpy as np
```

```python
# 기본차트
plt.title('plot1')
plt.plot([1, 3, 4, 9, 10, 17])
plt.show
```

```python
# x축 설정
plt.title('plot2')
plt.plot([10, 20, 30, 40, 50, 60], [1, 3, 4, 9, 10, 17])
plt.show

```

```python
# 차트 스타일
plt.title('plot3')
plt.plot([10, 20, 30, 40, 50, 60], [1, 3, 4, 9, 10, 17], 'go:') # x, y, 속성
plt.grid(True)

plt.show
```

```python
# 멀티 라인차트
values = np.arange(0, 5, 0.2)
print(values)

plt.title('plot4')
plt.plot(values, values, 'r--')
plt.plot(values, values ** 2, 'g--') # 2제곱
plt.plot(values, values ** 3, 'g--') # 3제곱


plt.show()
```

```python
# 막대차트
plt.title('plot5')
plt.bar(['apple', 'banana', 'orange', 'grape'], [5, 3, 7, 9])

plt.show
```

```python
# 히스토그램
dataset1 = np.random.rand(1000) # 0~1사이의 임의의 실수 1000개
dataset2 = np.random.randn(1000) # 표준정규분포를 따르는 임의의 실수 1000개

# dataset1
plt.title('Hist1')
plt.hist(dataset1, edgecolor='black')
plt.show
```

```python
# dataset1
plt.title('Hist2')
plt.hist(dataset2, edgecolor='black')
plt.show
```

```python
# 산점도
x = np.random.normal(0, 1, 1000)
y = np.random.normal(0, 1, 1000)

plt.title('Scatter')
plt.scatter(x, y)
plt.show
```

```python
# 산점도 3차원
import plotly.express as px
```
